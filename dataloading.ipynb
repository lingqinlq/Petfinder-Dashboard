Here we only want the movie after the Year 2000

```python
title_filtered = title_basics.copy()
title_filtered = title_filtered[title_filtered['titleType'].isin(['movie'])]
title_filtered = title_filtered[title_filtered['startYear'] != '\\N']
title_filtered['startYear'] = title_filtered['startYear'].astype('int64')
title_filtered = title_filtered[(title_filtered['startYear'] > 2000) & (title_filtered['startYear'] <= 2020)]
```
